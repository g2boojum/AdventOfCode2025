```python
testlines = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82""".splitlines()
```

```python
with open('day1input.txt') as fp:
    data = fp.read().splitlines()
```

## Part 1

```python
def part1(lines):
    pos = 50
    count = 0
    modulus = 100
    direc = {'L': -1, 'R': +1}
    for cmd in lines:
        pos = (pos + direc[cmd[0]]*int(cmd[1:])) % modulus
        if pos == 0:
            count += 1
    return count
```

```python
assert(3 == part1(testlines))
```

```python
part1(data)
```

## Part 2

```python
def part2(lines):
    pos = 50
    count = 0
    modulus = 100
    direc = {'L': -1, 'R': +1}
    for cmd in lines:
        mult = direc[cmd[0]]
        delta = int(cmd[1:])
        full_spins = delta // modulus
        count += full_spins
        delta -= full_spins*modulus
        newpos = (pos + mult*delta) % modulus
        if newpos == 0:
            count += 1
        elif (pos != 0) and (mult == -1) and (newpos > pos):
            # turning to the left, but newpos > pos means passed 0
            count += 1
        elif (pos != 0) and (mult == +1) and (newpos < pos):
            # turning to the right, but newpos < pos means passed 0
            count += 1
        pos = newpos
    return count
```

```python
assert(6 == part2(testlines))
```

```python
part2(data)
```
